```python
# Finding the csv from the league
# Multi League directory
country = 'ES'
league = '140'
season = 2016
week = 38
dirName = '../' + country + '/' + league + '/' + str(season) + '/'
# csv where save
file_title = str(season) + '_' + str(week) + '.csv'
csv_file = dirName + file_title
df = pd.read_csv(csv_file)
df.head(5)
```

```python
df = df.drop(df.columns[0], axis=1)
df.head(5)
```

```python
i_th = df.columns.get_loc("team_home.points")
i_ta = df.columns.get_loc("team_away.points")
i_th = df.columns.get_loc("team_home.points")
i_ta = df.columns.get_loc("team_away.points")
i_w = df.columns.get_loc("week")
for i in range(len(df)):
    for index, row in df.iterrows():
        # getting just current season
        if df.iloc[i,i_w] == 1:
            df.iloc[i,i_th] = 0
            df.iloc[i,i_ta] = 0
```

```python
barca_home = df[(df["team_home.team.id"]==529)]
barca_away = df[(df["team_away.team.id"]==529)]
barca = barca_home.append(barca_away)
barca.head(5)
```

```python
def home_points(points, goals_home, goals_away):
    if goals_home > goals_away:
        return points + 3
    if goals_home == goals_away:
        return points + 1
    else
        return points
```

```python
def away_points(points, goals_home, goals_away):
    if goals_home < goals_away:
        return points + 3
    if goals_home == goals_away:
        return points + 1
    else
        return points
```

```python
for w in range(2, 4):
    if len(barca_home[(barca_home["week"] == w)]) > 0:
        barca_home[(barca_home["week"] == w)]["team_home.points"].apply(lambda row : home_points(row['team_home_points'], row['goals_home'], 
                                  row['goals_away']), axis = 1) 
        print("Home game")
        print(w)
    else:
        print("Away game")
        print(w)
        barca_away[(barca_away["week"] == w)]["team_away.points"].apply(lambda row : home_points(row['team_away_points'], row['goals_home'], 
                                  row['goals_away']), axis = 1) 
```
